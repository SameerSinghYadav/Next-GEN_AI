```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

sentences = [
    "i am learning python and python",
    "we are happy today",
    "i am sad today",
    "this is basically a type of data"
]

def preprocess_text(documents):
    Cleaned_data = []

    english_stopwords = stopwords.words("english")
    wnet = WordNetLemmatizer()

    for doc in documents:
        filtered_tokens = []
        final_tokens = []

        raw_text = doc.lower()
        print("After lower case : ", raw_text)

        tokens = word_tokenize(raw_text)
        print("Tokens:", tokens)

        for word in tokens:
            if word not in english_stopwords:
                filtered_tokens.append(word)

        print("Filtered Tokens :", filtered_tokens)

        punctuations = string.punctuation
        clean_tokens = [word for word in filtered_tokens if word not in punctuations]
        print("After removing punctuations : ", clean_tokens)

        lemmatized_words = []
        for word in clean_tokens:
            lemmatized_words.append(wnet.lemmatize(word, "v"))
        print("After lemmatization : ", lemmatized_words)

        for word in lemmatized_words:
            if word.isalpha():
                final_tokens.append(word)

        print("Final tokens : ", final_tokens)

        cleaned_text = " ".join(final_tokens)
        print("Cleaned Text : ", cleaned_text)

        Cleaned_data.append(cleaned_text)
        print("*" * 70)

    return Cleaned_data
cleaned_data = preprocess_text(sentences)
print("\nFinal cleaned data list:")
print(cleaned_data)
```
